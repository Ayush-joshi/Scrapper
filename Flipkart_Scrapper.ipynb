```python
from bs4 import BeautifulSoup
import requests

searchQuery = "Laptop" #We can pass any keyword
pageNumber = "1" #We can parse any number of pages
flipkartBaseUrl = "https://www.flipkart.com/search?q=" + searchQuery + "&page=" + pageNumber

def toGetListOfProductUrl(flipkartBaseUrl):
    # opening our url file in append mode
    #File = open("url.csv", "a")
 
    # specifying user agent, You can use other user agents
    # available on the internet
    HEADERS = ({'User-Agent':
                'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko)Chrome/44.0.2403.157 Safari/537.36',
                                'Accept-Language': 'en-US, en;q=0.5'})
 
    # Making the HTTP Request
    webpage = requests.get(flipkartBaseUrl, headers=HEADERS)
 
    # Creating the Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")
    try:
        # Outer Tag Object
        productDivs = soup.find_all('div',class_= "_13oc-S")
        #print(productDivs)
        for productDiv in productDivs:
            name = productDiv.find("div",class_="_4rR01T").text
            price = productDiv.find("div",class_="_30jeq3 _1_WHN1").text.strip().replace(",",'')
            rating= productDiv.find('div', class_='_3LWZlK').text
            reviews = productDiv.find('span',class_='_2_R_DZ').text
            a = productDiv.find("a",class_="_1fQZEK")
            url = "https://www.flipkart.com" + a.get("href")
            print(name)
            print(price)
            print(rating)
            print(reviews)
            print(url)
           # File.write(f"{a['href']},")
            
 
    except AttributeError:
        print("Not Available")
    
    #File.close()
```

```python
toGetListOfProductUrl(flipkartBaseUrl)
```
